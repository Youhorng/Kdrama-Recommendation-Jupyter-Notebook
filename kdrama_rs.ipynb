```python
movies = pd.read_csv('cleaned_data.csv')
movies
```

```python
movies.movies = movies.drop(columns=['Unnamed: 0'])
movies = movies[['Name', 'Genre', 'Sinopsis', 'Tags']]
movies
```

```python
movies.loc[:, 'Tags'] = movies['Tags'] + movies['Genre']
```

```python
movies.loc[movies['Name'] == 'Flower of Evil', 'Tags'].values[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 1623, stop_words='english')
```

```python
vector = cv.fit_transform(movies['Tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
movies.loc[movies['Name'] == 'Squid Game']
```

```python
distance = sorted(list(enumerate(similarity[132])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(movies.iloc[i[0]].Name)
```
